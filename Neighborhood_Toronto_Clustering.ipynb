### Loading Toronto Neighborhoods table from .ipnb file  Neighborhood_in_Toronto-geographical_coordinates.ipynb.

```python
dft1 = pd.read_csv('Toronto_Neighborhoods_GC.csv')
dft1
```

##### Using a geolocator to find a longitude and latitude of a main zone - Toronto city. 

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

##### Visualization of a Toronto using above coordinates map with zoom 10

```python
Toronto_N = folium.Map(location=[latitude, longitude], zoom_start=10) # generate Toronto map 
Toronto_N
```

##### Adding neighborhoods to Toronto Map using folium.

```python
# create map of Toronto using latitude and longitude values
map_newyork = folium.Map(location=[latitude, longitude], zoom_start= 10)

# add markers to map
for lat, lng, borough, neighborhood in zip(dft1['Latitude'], dft1['Longitude'], dft1['Borough'], dft1['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=6,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
   
map_newyork
```

##### Observations: Downtown Toronto has the most condensation of neighborhoods. We can notice that the rest of the neighborhoods are more spread out.
